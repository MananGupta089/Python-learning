```python
df['Parking'] = df['Parking'].fillna(df['Parking'].median())
df['Furnishing'] = df['Furnishing'].fillna(df['Furnishing'].mode()[0])
df['Type'] = df['Type'].fillna(df['Type'].mode()[0])
df['Bathroom'] = df['Bathroom'].fillna(df['Bathroom'].median())
df['Per_Sqft'] = df['Per_Sqft'].fillna(df['Per_Sqft'].median())
```

```python
one_hot_encoding = df.copy()
one_hot_encoding = pd.get_dummies(df,columns=['Type','Furnishing','Status','Transaction'],drop_first=True)
one_hot_encoding.drop(columns=['Locality'],inplace = True)
```

```python
from sklearn.preprocessing import StandardScaler , LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error


model = LinearRegression()
scalar = StandardScaler()
label = LabelEncoder()

X = one_hot_encoding.drop(columns='Price',axis=1)
Y = one_hot_encoding['Price']

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.20,random_state = 42)
X_train = scalar.fit_transform(X_train)
X_test = scalar.transform(X_test)

model.fit(X_train,Y_train)
y_pred = model.predict(X_test)

r2 = r2_score(Y_test,y_pred)
mse = mean_squared_error(Y_test,y_pred)
rmse = np.sqrt(mse)

print("Value of R2 is :",r2)
print("Value of MSE :",mse)
print("Value of RMSE :",rmse)



```
